### Inputan Fill Forward

```python
# Fill Forward
# feel of forwad pill
# Akan di dari nilai sebelum nya
# Kelemehan nya adalah jika nilai index sebelum nya kosong maka
# nilainya jug akan kosong
df.fillna(method='ffill')
```

### Inpulation Fill Mean

```python
# Fill mean / Rata-Rata
# df.fillna(value=df.mean())
```

```python
# Memperbaiki mean secara spetifik yaitu target nya
# Menyatu fungsi rata-rata yang hanya numeric only saja
# kalau masih ada nan di object di karenakan yang di isi adalah numeric saja
df.fillna(value=df.mean(numeric_only=True))
```

### Imputation fill Median

```python
# Fill median
# seluruh gaji
# median pada MasaKerja adalah 8
df.fillna(value=df.median(numeric_only=True))
```

```python
# cara-cara tadi hanya untuk mengisi non kategorikal atau bukan yang nominal
```

### Inputan Categorical

```python
# Fill categorical / nominal dengan Mode
# memunculkan .mode
# Mengambil kemunculan data yang terbenyak
df.fillna(df.select_dtypes(include='object').mode().iloc[0])
```

### Inputan Numiric & Categorical

```python
# Fill Atribut Numeric Mean, Atribut categorical Mode, Replace Data
# inplace=True mengganti data yang kosong tadi dengan dataset baru 
# yang sudah diisikan 
df.fillna(df.select_dtypes(include='object').mode().iloc[0], inplace=True)
df.fillna(df.select_dtypes(include='number').mean().iloc[0], inplace=True)
```
